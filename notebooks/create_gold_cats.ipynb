```python
import pandas as pd
import numpy as np
from datasets import Dataset
import os
```

```python
path_root = '../data/test_task/pooled/'

all_embs = Dataset.load_from_disk(f'{path_root}merged_all')
all_embs = all_embs.to_pandas()
```

```python
# create subset of newspapers we know for sure have international, national news, and advertisements

newspapers = all_embs['newspaper'].unique()

list_dk = ['Aalborg Stiftstidende', 'Aalborgs Stifts Adresseavis',
       'Aarhuus Stifts-Tidende', 'Berlingske Tidende', 'Den Nord-Cimbriske Tilskuer',
       'Københavns Adresseavis', 'Lolland-Falsters Stifts-Tidende',
       'Odense Adresse-Contoirs Efterretninger', 'Ribe Stifts Adresseaviser', 'Viborger Samler']

list_dk_no = ['Aalborg Stiftstidende', 'Aalborgs Stifts Adresseavis',
       'Aarhuus Stifts-Tidende', 'Berlingske Tidende', 'Den Nord-Cimbriske Tilskuer',
       'Københavns Adresseavis', 'Lolland-Falsters Stifts-Tidende',
       'Odense Adresse-Contoirs Efterretninger', 'Ribe Stifts Adresseaviser', 'Viborger Samler',
       'Efterretninger fra Adresse-Contoiret i Bergen', 'Tronhiems Adresse-Contoirs Efterretninger']

extra = ['Den Vest-Sjællandske Avis', 'Jyske Efterretninger', 'Norske Intelligenssedler']

subset_dk = all_embs[all_embs['newspaper'].isin(list_dk)]

subset_dk_no = all_embs[all_embs['newspaper'].isin(list_dk_no)]

subset_extra = all_embs[all_embs['newspaper'].isin(extra)]
```

```python
new_ds = Dataset.from_pandas(subset_dk.reset_index(drop=True))
save_path = os.path.join(path_root, 'subset_dk')
new_ds.save_to_disk(save_path)
```

```python
new_ds = Dataset.from_pandas(subset_dk_no.reset_index(drop=True))
save_path = os.path.join(path_root, 'subset_dk_no')
new_ds.save_to_disk(save_path)
```

```python
new_ds = Dataset.from_pandas(subset_extra.reset_index(drop=True))
save_path = os.path.join(path_root, 'subset_extra')
new_ds.save_to_disk(save_path)
```

### Load the predicted categories and take a sample

```python
subset_dk = pd.read_csv(f'{path_root}subset_dk_clean_predicted.csv', index_col=0)

subset_dk_no = pd.read_csv(f'{path_root}subset_dk_no_clean_predicted.csv', index_col=0)

subset_extra = pd.read_csv(f'{path_root}subset_extra_clean_predicted.csv', index_col=0)
```

```python
# take random sample of 100 per newspaper
sampled_extra = (
    subset_extra.groupby("newspaper", group_keys=False).apply(
    lambda x: x.sample(n=min(len(x), 100), random_state=42)
)
)
```

```python
sampled_extra.to_csv(f'{path_root}subset_extra_predicted_sample.csv')
```

```python
# take random sample of 100 per newspaper
sampled_dk = (
    subset_dk.groupby("newspaper", group_keys=False).apply(
    lambda x: x.sample(n=min(len(x), 100), random_state=42)
)
)

# take random sample of 100 per newspaper
sampled_dk_no = (
    subset_dk_no.groupby("newspaper", group_keys=False).apply(
    lambda x: x.sample(n=min(len(x), 100), random_state=42)
)
)
```

```python
sampled_dk.to_csv(f'{path_root}subset_dk_predicted_sample.csv')
sampled_dk_no.to_csv(f'{path_root}subset_dk_no_predicted_sample.csv')
```
